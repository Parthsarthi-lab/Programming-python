## PROBLEM-4> Take any two numbers( 2 digit +ve integers) as input from the user. Multiply the numbers in this way->


```python
print("Enter two numbers(2-digit +ve integers) for multiplication")
a= int(input())
b= int(input())

tens_a = a//10
ones_a = a % 10

tens_b = b//10
ones_b = b % 10

product = (ones_a*tens_b) + (tens_a*ones_b)

print("The final product is ",product)
```
